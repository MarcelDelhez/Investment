# VaR calculation

```python
import numpy as np
import pandas as pd
import yfinance as yf
from tabulate import tabulate
from datetime import date

# Plotting
import matplotlib.pyplot as plt
import seaborn
import matplotlib.mlab as mlab

#Statistical calculation
import math
from scipy.stats import norm

# We will import the daily data of Amazon from yahoo finance
# Calculate daily returns
AMZN = yf.download("AMZN", "2020-01-01", "2023-02-01")
df = AMZN[['Adj Close']]
df.insert(loc=1, column='returns', value=df['Adj Close'].pct_change())
df = df.dropna()
```

# VaR on a given period of time
To calculate the maximum loss over a given number of days, just multiply by square root of number of days.

# Calculate CVaR

```python
CVaR_90 = np.nanmean(df.loc[df['returns']<VaR_90]['returns'])
CVaR_95 = np.nanmean(df.loc[df['returns']<VaR_95]['returns'])
CVaR_99 = np.nanmean(df.loc[df['returns']<VaR_99]['returns'])

print (tabulate([['90%', VaR_90, CVaR_90], ['95%', VaR_95, CVaR_95], ['99%', VaR_99, CVaR_99]], headers = ['Confidence Level', 'VaR', 'CVaR']))
```

# What is practically mean.
Both VaR and CVaR aims to calculate the potential losses of an investment.
VaR gives the possible loss with a given percentage of certitude but do not takes care of what happens if it goes wronger than that.
CVaR does the same but takes the worsts cases in consideration.  It represents the mean losses that are worst than the VaR.
