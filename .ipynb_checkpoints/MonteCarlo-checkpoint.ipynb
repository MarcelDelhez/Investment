# Monte Carlo
Monte Carlo forecasting on S&P 500

```python
import numpy as np
import pandas as pd
from scipy.stats import norm
from numba import njit
import yfinance as yf

# Plotting
import matplotlib.pyplot as plt
import seaborn
from IPython.display import display, HTML
```

```python
# Retrieve stock prices
stocks=['^GSPC']
stocks.sort()
price = yf.download(stocks, "2018-01-01", "2023-02-05")['Adj Close']
```

```python
@njit(fastmath=True)
def Monte_Carlo(output, S0, mu, sigma, N_STEPS, N_PATHS, d_normals, IntStep):
    drift=(mu-0.5*sigma*sigma)
    for i in range(N_PATHS):
        S=0.0
        #mdd=0.0
        #maxPrev=S0
        W=0.0
        for k in range(1, N_STEPS+1):
            index=(k-1)+(i*N_STEPS)
            W+=d_normals[index]
            S=S0*np.exp((drift*k)+(sigma*W))
            if (k % IntStep)==0:
                output[i, int(k/IntStep)-1] = S
            #mdd=np.fmax(mdd,(maxPrev-S)/maxPrev)
            #maxPrev=np.fmax(maxPrev,S)
        output[i,-1]=S
        
def get_Monte_Carlo(S0, mu, sigma, seed=0, N_STEPS=1000, N_PATHS=1000, Interim=5):
    np.random.seed(seed)
    random_d=np.random.normal(0,1,N_PATHS*N_STEPS)
    IntStep=int(N_STEPS/Interim)
    output=np.zeros([N_PATHS, Interim], dtype=np.float32)
    Monte_Carlo(output, S0, mu, sigma, N_STEPS, N_PATHS, random_d, IntStep)
    return output
```

```python
N_Hist = 535
#N_Hist = len(price)
Sp = 0

price_N = price[Sp:Sp+N_Hist]
S0=price_N[-1]
dt=1
T=500
N=int(T/dt)
t=np.arange(1, int(N)+1)
pct=price_N.pct_change()
mu=np.log1p(pct).mean()
sigma=np.log1p(pct).std()
```

```python
Interim = int(T/10)
mc= get_Monte_Carlo(S0, mu, sigma, 0, int(N), 800, Interim)
```

```python
from scipy.optimize import curve_fit

@njit(fastmath=True)
def objective (x, a, b, c):
    return a*x + b*x*x + c

x_line = range(len(price_N))
y_line = price_N
plt.figure(figsize=(16, 9), dpi=92)
plt.title('Possible evolution for S&P 500.')
plt.plot(x_line, y_line, color='blue')
palP = seaborn.color_palette('YlGn_r')
palN = seaborn.color_palette('YlOrRd_r')

y_line=price[Sp+N_Hist:]
x_line=np.arange(0, len(y_line), 1)+N_Hist
plt.plot(x_line, y_line, color='palegreen')
x_line=np.arange(0, N, 1)+N_Hist
xr_line = np.concatenate((np.array(range(0, N+1, int(N/Interim))), np.array([N])), axis=0)[0:mc.shape[1]+1] + N_Hist
for i in [5, 10, 25, 50, 75, 90, 95]:
    yr_line = np.concatenate((np.array([price_N[-1]]), np.percentile(mc,  i, axis=0)), axis=0)
    popt, _ = curve_fit(objective, xr_line, yr_line)
    a, b, c = popt
    y_line=objective(x_line, a, b, c)
    plt.plot(x_line, y_line, color='red')
    plt.text(x_line[-1],y_line[-1],f"{100-i}% chances \n return > {(y_line[-1]/yr_line[0]-1)*100:3.1f}%")
plt.show()
```
